```python
import pandas as pd
import numpy as np
import seaborn as sea
```

```python
data = pd.read_excel("BlueEleData.xlsx", sheet_name=1)
data
```

```python
redCurry = data[data["Item Desc 1"].str.lower() == "red curry sauce"]
greenCurry = data[data["Item Desc 1"].str.lower() == "green curry sauce"]
yellowCurry = data[data["Item Desc 1"].str.lower() == "yellow curry sauce"]

yellowCurry
```

```python
redCurry_sorted = redCurry.sort_values(by="City", ascending=True)
city_sales_red = redCurry_sorted.groupby(["City", "State"])[["Range 1 POS Sales", "Range 1 POS Qty", "Range 2 POS Sales", "Range 2 POS Qty"]].sum().sort_values(by="Range 1 POS Sales", ascending=False)
city_sales_red.head(8)
```

```python
greenCurry_sorted = greenCurry.sort_values(by="City", ascending=True)
city_sales_green = greenCurry_sorted.groupby(["City", "State"])[["Range 1 POS Sales", "Range 1 POS Qty", "Range 2 POS Sales", "Range 2 POS Qty"]].sum().sort_values(by="Range 1 POS Sales", ascending=False)
city_sales_green.head(8)
```

```python
yellowCurry_sorted = yellowCurry.sort_values(by="City", ascending=True)
city_sale_yellow = yellowCurry.groupby(["City", "State"])[["Range 1 POS Sales", "Range 1 POS Qty", "Range 2 POS Sales", "Range 2 POS Qty"]].sum().sort_values(by="Range 1 POS Sales", ascending=False)
city_sale_yellow.head(8)
```
